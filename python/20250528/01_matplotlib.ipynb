```python
# 라이브러리 로드
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 라인 그래프 생성
# plot() 함수 이용
# 인자가 1개 -> y축에 데이터 대입
plt.plot([10, 3, 5, 1])
# show() 함수: print() 함수와 흡사한 기능 (출력)
plt.show()
```

```python
# plot() 함수
# 인자가 2개 -> 첫번째 인자는 x축, 두번째 인자는 y축에 대입
plt.plot(
    [1, 3, 7, 15],      
    [10, 3, 5, 1]
)
plt.show()
```

```python
# 선의 색상이나 종류 변경 -> 마지막 인자에 추가
plt.plot(
    [1, 2, 3, 4],
    [10, 5, 3, 1],
    '--r'
)
```

```python
plt.plot(
    [1, 2, 3, 4],
    [5, 7, 8, 10],
    "#00FF44"
)
plt.show()
```

```python
# x축, y축 그래프에 이름 붙이기
plt.plot(
    [1, 2, 3, 4],
    [10, 20, 30, 40]
)
plt.xlabel('name')
plt.ylabel('score')
plt.title('graph')
plt.show()
```

```python
# x, y 이름을 한글로 지정하는 경우
# 한글 깨짐 현상 발생 (matplotlib에서 기본으로 사용하는 폰트가 한글 미지원)
# 해결 -> matplot의 폰트설정 변경

plt.plot(
    [1, 2, 3, 4],
    [10, 20, 30, 40]
)
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('그래프')
plt.show()
```

```python
x_list = [1, 2, 3, 4]
y_list = [10, 20, 30, 40]
y2_list = [40, 30, 20, 10]
plt.plot(x_list, y_list)
plt.plot(x_list, y2_list)

# 범례 추가
# 기본 위치: 선을 가리지 않는 최적의(best) 위치
# 위치 고정하려면 -> loc 매개변수에 위치 지정 
# 숫자 or 문자로 
plt.legend(['A', 'B'], loc='upper center')
plt.show()
```

```python
# 막대 그래프 => bar()
plt.bar(x_list, y_list)
plt.show()
```

```python
# 수평 막대 그래프 => barh()
plt.barh(x_list, y_list)

plt.show()
```

```python
# 막대 색상 변경, 너비 변경
plt.bar(x_list, y_list, color=['r', 'g'], width=0.4)       # 막대 수보다 색상 지정 수가 적으면 해당 색 반복

plt.show()
```

```python
# 특정 막대의 색상 변경
graph = plt.bar(x_list, y_list)

graph[2].set_color('g')
plt.show()
```

```python
# 데이터 로드
drinks = pd.read_csv('../../csv/drinks.csv')
```

```python
drinks.head()
```

```python
plt.scatter(
    drinks['beer_servings'],
    drinks['spirit_servings']
)
plt.show()
```

```python
# pandas 라이브러리 안에 있는 상관계수 출력 함수
# corr()
# method 매개변수: 상관계수를 구하는 방법 설정
# drinks 안의 beer_servings 부터 total_litres_of_pure_alcohol 컬럼까지의 상관계수 확인

# 원하는 열만 추출 -> 5가지 방법
drinks.drop(['country', 'continent'], axis=1)
drinks[ ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']]
drinks.loc[:, 'beer_servings':'total_litres_of_pure_alcohol']
drinks.iloc[:, 1:5]
drinks.iloc[:, 1:-1]
```

```python
drinks_corr = drinks.iloc[:, 1:-1].corr(method='pearson')
```

```python
# 히트맵 출력
cols_view = ['beer', 'spirit', 'wine', 'alcohol']

hm = sns.heatmap(
    drinks_corr.values,         # 시각화할 데이터들
    xticklabels= cols_view,     # x축 label
    yticklabels= cols_view,     # y축 label
    annot = True,        # 히트맵에 숫자를 표기할것인가
    annot_kws= {'size' : 20},    # 숫자 크기를 조정
    fmt = '0.3f',        # 소수점 세 자리만 출력
    cmap = 'Greens',     # 히트맵에 기준이 되는 색상 지정 (상관계수 높을수록 지정 색에 가까워짐)
    square = False,       # 히트맵의 사각형을 정사각형으로 표시할 것인가
    cbar = True         # 사이드바를 표시할 것인가
)

plt.show()
```

```python
# 데이터의 분포도 확인
sns.pairplot(
    drinks.iloc[ :, 1:-1 ]
    # corner = True       분포도를 반만 보여줌
)
plt.show()
```

```python
import pandas as pd
```

```python
# csv 폴더에 있는 uriage.csv 파일을 로드
uriage = pd.read_csv('../../csv/uriage.csv')
```

```python
# 반복문 이용해 item_name 컬럼의 데이터들을 모두 대문자로 변경
# 비어있는 리스트 하나 생성
# 반복실행하면서 대문자로 변환된 값을 비어있는 리스트에 추가
item_names = []
# for문
for i in range(len(uriage)):
    data = uriage.loc[i, 'item_name'].upper()
    item_names.append(data)
# uriage.item_name = item_names     item_names를 uriage의 item_name에 대입

print(len(item_names))
print(item_names)
```

```python
df = uriage.copy()
df.item_name = item_names
```

```python
uriage['item_name'] = uriage['item_name'].str.upper()               # 대문자 변환
uriage['item_name'] = uriage['item_name'].str.replace(' ', '')      # 공백 제거
```

```python
# 반복 실행할 코드 작성
# df2의 item_name이 '상품A'이고, item_price가 결측치가 아닌 데이터에서
# item_price의 평균값을 구한다

flag_null = df2['item_price'].isna()        # item_price가 결측치인 조건
flag = df2['item_name'] == '상품A'          # item_name이 '상품A'인 조건

mean_price = df2.loc[
    flag & ~flag_null,
    'item_price'            # 컬럼의 조건이 문자 -> Series 형태 반환
    # 'item_price' 대신 ['item_price'] 사용 시 -> DataFrame 형태 반환
].mean()

mean_price
```

```python
flag_null = df2['item_price'].isna()

for i in range(len(df2['item_name'].unique())):
    flag = df2['item_name'] == df2['item_name'].unique()[i]

    mean_price = df2.loc[
        flag & ~flag_null,
        'item_price'
    ].mean()
    df2.loc[flag & flag_null, 'item_price'] = mean_price
    
```

1. for 반복문

```python
# 반복문 이용해 item_name 컬럼의 데이터들을 모두 대문자로 변경
# 비어있는 리스트 하나 생성
# 반복실행하면서 대문자로 변환된 값을 비어있는 리스트에 추가
item_names = []
# for문
for i in range(len(uriage)):
    data = uriage.loc[i, 'item_name'].upper()
    item_names.append(data)
# uriage.item_name = item_names     item_names를 uriage의 item_name에 대입

print(len(item_names))
print(item_names)
```

2. while 반복문

```python
item_names = []
# while문
# 초기값 (시작값=0)
i = 0
# while 조건식 (i<3000)
while i<2999:
    data = uriage.loc[i, 'item_name'].upper()
    item_names.append(data)
    i += 1
    
print(len(item_names))
print(item_names)
```

```python
uriage.head()
```

3. while True: 반복문

```python
item_names = []
i = 0
while True:
    data = uriage.loc[i, 'item_name'].upper()
    item_names.append(data)
    
    if i == len(uriage) - 1:
        break
    i += 1
print(len(item_names))
print(item_names)
```

```python
uriage.head()
```

4. try ~ except문

```python
# 특정 코드를 실행하고 만약 에러 발생시 예외 처리
# try ~ except 문
item_names = []
i = 0

while True:
    # 코드 실행을 시도
    try:
        data = uriage.loc[i, 'item_name'].upper()
        item_names.append(data)
        i += 1
        # i가 2998까지는 정상 작동
        # i가 2999가 되면 에러 발생
    except:
        # 위의 try code 에러 발생 시 실행 부분
        break
print(len(item_names))
print(item_names)
```

```python
# 데이터프레임에서 특정 컬럼의 데이터를
# 대문자로 변환하고 문자열 사이에 존재하는 공백을 모두 제거해서
# 데이터를 변환해주는 함수 생성

# 매개변수 -> 함수 안에서만 사용 가능한 변수 (입력 데이터)
# 2개 -> 데이터프레임, 특정 컬럼
def test_func(_df, _col):
    # _df : 데이터프레임
    # _col : 데이터를 변경할 컬럼의 이름
    # _df의 복사본 생성
    df = _df.copy()
    df[_col] = df[_col].str.lower().str.replace(' ', '')
    return df
```

```python
# item_name 데이터 추출
unique_data = []
for item in df2['item_name']:
    # print(i)
    # item이 unique_data에 포함되어 있지 않다면
    if item not in unique_data:
        unique_data.append(item)
unique_data
```

```python
# 반복문을 이용해 데이터프레임의 item_price 결측치 채우기
# 반복 횟수: item_name의 유니크한 데이터 개수만큼
item_list = df2['item_name'].unique()

# item_price가 결측치인 조건 생성
flag_null = df2['item_price'].isna()

# item_list를 이용해 반복문 실행
for item in item_list:
    # item에는 item_list의 원소가 하나씩 대입
    # print(item)
    # 조건식 생성: df2에서 item_name이 item과 같다
    flag = df2['item_name'] == item
    # flag와 flag_null의 부정이 모두 만족하는 경우의 item_price의 평균
    mean_price = df2.loc[flag & ~flag_null, 'item_price'].mean()
    # flag와 flag_null을 모두 만족하는 경우의
    # item_price에 mean_price를 대입
    df2.loc[flag & flag_null, 'item_price'] = mean_price
```

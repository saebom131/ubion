```python
import pandas as pd
```

```python
corona = pd.read_csv('../../csv/corona.csv')
```

```python
corona.drop(['Unnamed: 0', 'seq'], axis=1, inplace=True)
corona.columns = ['등록일시', '총사망자', '총확진자', '기준일', '기준시간', '수정일시', '누적의심지', '누적확진율']
```

```python
df = corona.copy()
```

```python
df.sort_values(['등록일시'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
datas = [0]

for i in range(1, len(df), 1):
    data = df.loc[i, '총사망자'] - df.loc[i-1, '총사망자']
    datas.append(data)
```

```python
df['일일사망자'] = datas
```

```python
df.head()
```

```python
df['일일사망자2'] = (df['총사망자']-df['총사망자'].shift()).fillna(0)
```

```python
df.head()
```

```python
df['일일사망자3'] = df['총사망자'].diff().fillna(0)
```

### <Sales Record.csv 파일>
1. 'Order Priority', 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit' 컬럼 제거
2. 컬럼의 이름을 변경: [대륙, 국가, 상품종류, 판매채널, 주문일자, 발송일자, 판매개수, 단가, 원가]
3. 대륙 -> 국가 기준으로 오름차순 정렬
4. 파생변수 생성
    - 총단가 = 단가 * 판매개수
    - 총원가 = 원가 * 판매개수
    - 총이윤 = 총단가 - 총원가

```python
# 몇몇 컬럼 제거 후 새로운 데이터프레임 'df'에 저장
df = sales.drop(['Order Priority', 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit'], 
           axis='columns')     # axis=1도 같은 의미
```

```python
df.head()
```

```python
# 컬럼의 이름을 변경
# 두 가지 방법 : df.columns(전체 컬럼명 변경 시 유리), df.rename(특정 컬럼명 변경 시 유리)
# df.columns = ['대륙', '국가', '상품종류', '판매채널', '주문일자', '발송일자', '판매개수', '단가', '원가']
df.rename(
    columns = {
        'Region' : '대륙',
        'Country' : '국가',
        'Item Type' : '상품종류',
        'Sales Channel' : '판매채널',
        'Order Date' : '주문일자',
        'Ship Date' : '발송일자',
        'Units Sold' : '판매개수',
        'Unit Price' : '단가',
        'Unit Cost' : '원가'
    }
)

df.columns = ['대륙', '국가', '상품종류', '판매채널', '주문일자', '발송일자', '판매개수', '단가', '원가']
```

```python
# 대륙 -> 국가를 기준으로 오름차순 정렬
df.sort_values(['대륙', '국가'], inplace=True)
```

```python
df.head()
```

```python
# 파생변수 생성: '총단가', '총원가'
df['총단가'] = df['단가'] * df['판매개수']
df['총원가'] = df['원가'] * df['판매개수']
```

```python
df.head()
```

```python
# 파생변수 생성: '총이윤'
df['총이윤'] = df['총단가'] - df['총원가']
```

```python
# sales의 상위 5개 데이터 확인
df.head()
```

```python
# 대륙을 기준으로 그룹화한 group_data의 총이윤의 평균
group_data[['총이윤']].mean()
```

```python
# pivot_table() 함수: 테이블 데이터를 재구조화
# index를 어떠한 기준으로 사용할 것인가
# column을 어떤 기준으로 사용할 것인가
# values는 어떤 데이터를 이용할 것인가
# 연산이 필요하다면 연산
# pd.pivot_table()
    # 라이브러리(pandas) 안에 있는 함수를 호출할 때는 데이터를 가지고 있지 않기 때문에
    # 어떠한 데이터를 변경할 것인가? 인자로 사용
# df.pivot_table()
    # 클래스(DataFrame) 안의 함수를 호출할 때는 클래스가 기존 데이터를 가지고 있기 때문에
    # 어떠한 데이터를 변경할 것인가는 인자로 사용X

# 라이브러리(pandas) 안의 함수 호출
pd.pivot_table(
    df,
    index = ['대륙'],
    values = ['총이윤'],
    aggfunc = 'sum'
)
```

```python
# 클래스(DataFrame) 안의 함수 호출
df.pivot_table(
    index = ['대륙'],
    values = ['총이윤'],
    aggfunc= 'sum'
)
```

```python
df.pivot_table(
    index = ['대륙', '상품종류'],       # 대륙별, 상품종류별 그룹화
    values= ['총이윤'],
    aggfunc='sum'
)
```

```python
df.pivot_table(
    index = ['대륙'],
    values= ['총이윤'],
    columns = ['상품종류'],             # 상품종류를 컬럼의 데이터로 사용
    aggfunc='sum'
)
```

```python
# 그룹화 -> 대륙, 상품종류
# 그룹화 연산 -> 총이윤의 합계
group_data = df.groupby( ['대륙', '상품종류'] )['총이윤'].sum()
```

```python
# 인덱스의 기준을 컬럼으로 이동
# unstack() 함수: 인덱스의 기준을 컬럼으로 이동
group_data.unstack()
```

```python
# 대륙, 상품종류를 기준으로 그룹화
# 그룹화 연산 -> 판매개수의 합계 
# 아시아에서 가장 많이 팔린 상품의 종류
```

```python
# case 1
# groupby() => DataFrame 형태
case1_data = df.groupby(['대륙', '상품종류'])[['판매개수']].sum()
# index가 Asia인 데이터만 추출
case1_data.loc['Asia', ].sort_values('판매개수', ascending=False)
```

```python
# case 2
# groupby() => Series 형태
case2_data = df.groupby(['대륙', '상품종류'])['판매개수'].sum()
case2_data = case2_data.unstack()
case2_data
```

```python
# Series의 sort_values() 함수는 기준 데이터 지정X, 내림차순/오름차순 선택만 하면 됨(기본-오름차순)
case2_data.loc['Asia', ].sort_values(ascending=False)
```

```python
# case 3
# pivot_table() 함수 이용
case3_data = df.pivot_table(
    index = '대륙',
    values = '판매개수',
    columns = '상품종류',
    aggfunc = 'sum'
)
case3_data
```

```python
case3_data.loc['Asia', ].sort_values(ascending=False)
```

#### 위의 코드들을 함수화
- 첫번째 함수
    - 매개변수 1개
        - driver: webdriver
    - table 태그를 찾아서 데이터프레임화
    - url 찾아서 데이터프레임에 대입
    - 데이터프레임을 리턴    

```python
def create_df(web):
    # 해당 페이지의 페이지소스를 이용해서 데이터프레임 생성
    # 페이지소스를 문자화
    html_data = str(web.page_source)
    # 문자화된 페이지 소스를 read_html() 인자로 사용
    # 결과의 첫번째 데이터프레임을 변수에 대입
    df = pd.read_html(html_data)[0]
    # 해당 페이지소스를 BeautifulSoup으로 파싱 
    soup = bs(web.page_source, 'html.parser')
    # soup에서 tbody 태그 중 id가 tbody인 태그를 선택 
    tbody_data = soup.find(
        'tbody', attrs={
            'id' : 'tbody'
        }
    )
    tr_list = tbody_data.find_all('tr')
    url_list = []   # 빈리스트
    for tr_data in tr_list:
        url = base_url + tr_data.find_all('a')[1]['href']
        url_list.append(url)
    # df에 url_list 새로운 컬럼(url)에 대입 
    df['url'] = url_list
    # 데이터프레임을 리턴
    return df
    
```

1. driver에서 class가 pageSkip 태그를 선택
2. li 태그들을 모두 찾는다. -> 3번째 부터 마지막에서 2번째 전까지데이터를 저장

```python
# 특정 페이지까지 테이블의 내용을 불러오는 함수

def dart_save(_name, _cnt=1):
    # 기본 주소를 변수에 대입
    base_url = 'https://dart.fss.or.kr/'
    driver = webdriver.Chrome()
    driver.get(base_url)
    time.sleep(1)
    # id가 textCrpNm2인 태그 선택
    input_tag = driver.find_element(By.ID, 'textCrpNm2')
    # input_tag에 특정 데이터 입력
    input_tag.send_keys(_name)
    # input_tag에 ENTER 이벤트 발생
    input_tag.send_keys(Keys.ENTER)
    time.sleep(1)

    # 비어있는 데이터프레임 생성
    result = pd.DataFrame()
    for idx in range(_cnt):
        # class가 pageSkip 태그를 선택
        skip_tag = driver.find_element(By.CLASS_NAME, 'pageSkip')
        # <<, <, >, >>가 있는 페이지의 경우 li_list 생성
        # skip_tag에서 li 모두 찾아서 3번째부터 마지막 2번째 전까지
        li_list = skip_tag.find_elements(By.TAG_NAME, 'a')[2:-2]
        

        li_list[idx+1].click()
        df = create_df(driver)
        # result와 df를 단순 행결합
        result = pd.concat( [result, df] , axis=0 )
        time.sleep(1)

    result.to_csv(f'{_name}.csv', index=False)
    print(f'{_name}.csv 파일 생성 완료')

    driver.close()      # 드라이버 종료
```

```python
dart_save('SK하이닉스', 3)      # 조회 페이지수가 10개 이상이어야 정상 작동
```

```python
df = pd.read_csv('삼성전자.csv')
for i in range(len(df)):
    rep_url = df.iloc[i, -1]    # df에서 접속할 보고서 URL 추출
    try:
        driver = webdriver.Chrome()
        driver.get(rep_url)
        time.sleep(1)

        # class가 rightWrap 태그 선택
        div_tag = driver.find_element(By.CLASS_NAME, 'rightWrap')
        # button 태그를 모두 찾아 뒤에서 3번째 버튼 태그 선택
        button_tag = div_tag.find_elements(By.TAG_NAME, 'button')[-3]
        button_tag.click()
        
        # driver의 창을 이동
        driver.switch_to.window(driver.window_handles[1])
        # class가 'btnFile'인 태그를 모두 찾는다
        btn_tags = driver.find_elements(By.CLASS_NAME, 'btnFile')
        for btn in btn_tags:
            btn.click()
            time.sleep(1)
        print(f'{i+1}번째 보고서 저장 완료')
        driver.quit()   # 열려있는 모든 창 닫기
    except:
        print(rep_url)
```

```python
df2 = pd.DataFrame('SK텔레콤.csv')
for url in list(df2['url']):
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(1)

    # class가 rightWrap 태그 선택
    div_tag = driver.find_element(By.CLASS_NAME, 'rightWrap')
    # button 태그를 모두 찾아 뒤에서 3번째 버튼 태그 선택
    button_tag = div_tag.find_elements(By.TAG_NAME, 'button')[-3]
    button_tag.click()
    
    # driver의 창을 이동
    driver.switch_to.window(driver.window_handles[1])
    # class가 'btnFile'인 태그를 모두 찾는다
    btn_tags = driver.find_elements(By.CLASS_NAME, 'btnFile')
    for btn in btn_tags:
        btn.click()
        time.sleep(1)
    print(f'{url} 파일 저장 완료')
    driver.quit()   # 열려있는 모든 창 닫기
```

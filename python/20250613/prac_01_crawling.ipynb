```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

1. webdriver를 오픈해서 'https://dart.fss.or.kr/'에 요청 전송
2. id가 'textCrpNm2'인 태그를 선택
3. 특정 입력값을 전송 (send_key()) -> ENTER 이벤트 발생
4. 해당 페이지는 table 태그가 1개 존재
    - 해당 페이지의 html 소스코드를 문자화
    - 문자형 데이터를 pandas에 있는 read_html() 함수에 인자로 사용
    - read_html() 결과에서 첫번째 데이터프레임을 변수에 저장
5. 페이지 소스코드를 BeautifulSoup으로 파싱
6. tbody 태그 중 id가 'tbody'인 태그 선택
7. tr 태그들을 모두 검색
8. 각 tr 태그에서 a 태그를 모두 검색하고, 두번째 a 태그의 href 속성값 추출
9. base_url과 href 속성값을 더해 새로운 리스트에 추가
 


```python
driver = webdriver.Chrome()
```

```python
driver.get('https://dart.fss.or.kr/')
```

```python
element = driver.find_element(By.ID, 'textCrpNm2')
```

```python
element.send_keys('삼성전자')
```

```python
element.send_keys(Keys.ENTER)
```

```python
html_data = driver.page_source
html_data
```

```python
print(type(html_data))
```

```python
dfs = pd.read_html(html_data)
dfs
```

```python
df = dfs[0]
df.head(2)
```

```python
soup = bs(html_data, 'html.parser')
```

```python
soup
```

```python
tbody_data = soup.find(
    'tbody', attrs={
        'id' : 'tbody'
    }
)
```

```python
len(tbody_data.find_all('tr'))
```

```python
tr_list = tbody_data.find_all('tr')
tr_list
```

```python
tr_list[0].find_all('a')
```

```python
tr_list[0].find_all('a')[1]['href']
```

```python
href_list = []
for tr_data in tr_list:
    url = tr_data.find_all('a')[1]['href']
    href_list.append(url)
```

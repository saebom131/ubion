## bs4 라이브러리
- bs4 라이브러리 안에 존재하는 BeautifulSoup Class 이용
- html로 이루어진 문자열 데이터(str)를 parsing 작업을 통해 타입 변경 -> 데이터를 쉽게 추출하기 위한 방법
- html에서 TAG(element)를 기준으로 데이터를 쉽게 추출
- 웹의 구조를 어느 정도 파악하고 사용하면 쉽게 접근 가능
- Parser를 활용해 python에서 접근하기 쉽게 객체(변수+함수)의 형태로 제공

```python
# # 라이브러리 설치
# !pip install bs4
```

```python
# 라이브러리 로드 -> bs4 안의 BeautifulSoup 사용
from bs4 import BeautifulSoup as bs
```

```python
type(html_text)
```

```python
# BeautifulSoup class는 생성자 함수에서
# 필수 매개변수 2개
    # html로 이루어진 문자열 데이터
    # Parser 지정
soup = bs(html_text, 'html.parser')
```

```python
soup
```

```python
type(soup)          # Parsing(데이터 변환) 완료
```

- 태그의 이름을 기준으로 검색(필터링)
    - soup.태그명: 해당 태그명의 첫번째 태그의 정보 출력
    - soup.태그명.string: 첫번째 태그에서 문자열을 출력
        - ex> `<p>test</p>` -> test 출력
    - soup.태그명['속성명']: 첫번째 태그의 해당 속성값을 출력
        - ex> <a href='http://www.google.com`>google</a> -> 'http://www.google.com'

```python
soup.div['id']
```

- find()
    - html 데이터에서 특정 태그의 첫번째 정보 출력
    - find(속성명 = 속성값) : 태그들 중 해당 속성명과 속성값을 가진 태그의 첫번째 정보 출력
    - 결과 -> TAG 데이터 타입 반환

- find_all()
    - html 데이터에서 특정 태그의 모든 정보 출력
    - find_all(속성명 = 속성값) : 태그들 중 해당 속성명과 속성값을 가진 태그의 모든 정보 출력
    - limit 매개변수 : 찾을 태그의 개수 지정
    - 결과 -> TAG들의 집합인 ResultSet 형태로 반환(TAG를 문자라고 생각하면, ResultSet은 리스트와 흡사)

```python
# # ResultSet의 형태라 속성값을 바로 추출하는 것이 불가능 -> 에러
# soup.find_all('a')['href']
```

```python
a_list = soup.find_all('a')     # a 태그들의 리스트
```

```python
a_list[1]['href']
```

```python
# ResultSet 안의 데이터를 반복문으로 추출 -> 태그의 형태로 나옴
for data in a_list:
    print(data)
    break
```

```python
# a_list에서 href의 속성 값들을 새로운 리스트에 추가
href_list = []

for data in a_list:
    href_list.append(data['href'])

href_list
```

```python
# map 함수 -> python 기본 내장, pandas에도 존재
# python에서 제공하는 map 함수 사용
# map( function, list )
# python의 map함수 결과 -> map data -> list 형태로 변환
list(
    map(
        lambda x : x['href'],
        a_list
    )
)
```

```python
def change(x):
    return x['href']

list(
    map(
        change,
        a_list
    )
)
```

### 서울시 공동주택 아파트 정보 API 받아오기
- https://data.seoul.go.kr/dataList/OA-15818/S/1/datasetView.do


```python
import requests
from dotenv import load_dotenv
import os
```

```python
# .env에 변수 등록
load_dotenv()
```

```python
service_key = os.getenv('service_key')
service_key
```

- params를 사용하지 않고 주솟값 내에 직접 데이터 타입, 값 넣는 방식

```python
# params 사용x
# 주솟값 내에서 데이터의 타입 변경 가능
data_type = 'json'
start_num = '1'
end_num = '5'
url = f'http://openapi.seoul.go.kr:8088/{service_key}/{data_type}/OpenAptInfo/{start_num}/{end_num}/'
# url = 'http://openapi.seoul.go.kr:8088/(인증키)/xml/OpenAptInfo/1/5/'
```

```python
# 특정 주소로 요청 보냄
res = requests.get(url)
# 주솟값 내에 데이터의 정보 들어있으므로 params 매개변수 필요X
```

```python
# requests를 이용해 들어온 데이터가 json 형태인 경우
import json
```

```python
dict_data = json.loads(res.content)
dict_data
```

## bs4 라이브러리
- bs4 라이브러리 안에 존재하는 BeautifulSoup Class 이용
- html로 이루어진 문자열 데이터(str)를 parsing 작업을 통해 타입 변경 -> 데이터를 쉽게 추출하기 위한 방법
- html에서 TAG(element)를 기준으로 데이터를 쉽게 추출
- 웹의 구조를 어느 정도 파악하고 사용하면 쉽게 접근 가능
- Parser를 활용해 python에서 접근하기 쉽게 객체(변수+함수)의 형태로 제공

```python
# # 라이브러리 설치
# !pip install bs4
```

```python
# 라이브러리 로드 -> bs4 안의 BeautifulSoup 사용
from bs4 import BeautifulSoup as bs
```

```python
type(html_text)
```

```python
# BeautifulSoup class는 생성자 함수에서
# 필수 매개변수 2개
    # html로 이루어진 문자열 데이터
    # Parser 지정
soup = bs(html_text, 'html.parser')
```

```python
soup
```

```python
type(soup)          # Parsing(데이터 변환) 완료
```

- 태그의 이름을 기준으로 검색(필터링)
    - soup.태그명: 해당 태그명의 첫번째 태그의 정보 출력
    - soup.태그명.string: 첫번째 태그에서 문자열을 출력
        - ex> `<p>test</p>` -> test 출력
    - soup.태그명['속성명']: 첫번째 태그의 해당 속성값을 출력
        - ex> <a href='http://www.google.com`>google</a> -> 'http://www.google.com'

- find()
    - html 데이터에서 특정 태그의 첫번째 정보 출력
    - find(속성명 = 속성값) : 태그들 중 해당 속성명과 속성값을 가진 태그의 첫번째 정보 출력
    - 결과 -> TAG 데이터 타입 반환

- find_all()
    - html 데이터에서 특정 태그의 모든 정보 출력
    - find_all(속성명 = 속성값) : 태그들 중 해당 속성명과 속성값을 가진 태그의 모든 정보 출력
    - limit 매개변수 : 찾을 태그의 개수 지정
    - 결과 -> TAG들의 집합인 ResultSet 형태로 반환(TAG를 문자라고 생각하면, ResultSet은 리스트와 흡사)

#### 네이버 증권 페이지에서 데이터 추출

```python
# 네이버 증권에 요청
finance_res = requests.get('https://finance.naver.com/')
```

```python
html_data = finance_res.text
```

```python
# BeautifulSoup을 이용한 데이터 parsing
finance_soup = bs(html_data, 'html.parser')
finance_soup.title      # 정상 parsing 확인 -> title 존재
```

```python
finance_soup
```

1. 오늘의 증시 데이터(코스피, 코스닥, 코스피200)

```python
# div 태그 중 class(속성명)가 article(속성값)인 태그 확인
len(
    finance_soup.find_all(
        'div', attrs={          # attrs: 속성
            'class' : 'article'
        }
    )
)
# 출력 1의 의미 -> div 태그 중 class가 article인 태그는 1개이다. (id값으로 데이터 뽑을 때는 길이 확인 불필요)
```

```python
# 출력 = 1이므로 find(), find_all() 함수를 이어서 사용하기 위해 find() 함수 이용해 Tag 데이터 생성
article_data = finance_soup.find(
    'div', attrs={
        'class' : 'article'
    }
)
article_data
```

```python
# article_data에서 div 중 class가 section2인 태그 찾기
section2_data = article_data.find(
    'div', attrs={
        'class' : 'section2'
    }
)
section2_data
```

```python
# section2_data에서 첫번째 div 찾기
div_data = section2_data.find('div')
div_data
```

```python
# div_data에서 div 중 class가 section_stock인 태그 선택
section_stock_data = div_data.find(
    'div', attrs={
        'class' : 'section_stock'
    }
)
section_stock_data
```

```python
# section_stock_data에서 모든 div 찾기
div_list = section_stock_data.find_all(
    'div', attrs={
        'class' : 'group_quot'
    })
len(div_list)
```

```python
div_list[0].find('h4').string
```

```python
div_list[0].find(
    'span', attrs={'class' : 'num_quot'}
).get_text().split()
```

```python
(div_list[0].find('h4').get_text() \
+ div_list[0].find(
    'span', attrs={'class' : 'num_quot'}
).get_text()).split()
```

```python
for i in range(len(div_list)):
    res = (div_list[i].find('h4').get_text() \
        + div_list[i].find(
                'span', attrs={'class' : 'num_quot'}
            ).get_text()).split()
    print(res)
```

```python
new_list = []

for data in div_list:
    h4_text = data.find('h4').get_text()
    span_text = data.find('span', attrs=
        {'class' : 'num_quot'}
    ).get_text()
    data_list = (h4_text + span_text).split()
    new_list.append(data_list)
    
new_list
```

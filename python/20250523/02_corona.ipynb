### <데이터프레임에서 정렬 변경>
- 정렬 -> sort
- 정렬을 바꾸는 기준 : index / values
    - sort_index() : index의 값을 기준으로 정렬 변경
    - sort_vlaues() : 특정 컬럼의 값들을 기준으로 정렬 변경


```python
# 데이터프레임에서 정렬 변경
# 정렬 -> sort
# 정렬을 바꾸는 기준 : index / values
# sort_index() : index의 값을 기준으로 정렬 변경
# sort_vlaues() : 특정 컬럼의 값들을 기준으로 정렬 변경
```

```python

# 등록일시를 기준으로 오름차순 정렬
# 차순 정렬 방법 매개변수 ascending (기본값=True => 오름차순)
df.sort_values(
    ['등록일시'],
    inplace=True
)
```

```python
df
```

```python
# 인덱스를 초기화
# 인덱스 -> index
# 초기화 -> reset
# reset_index() 함수 사용
df.reset_index()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
# 일일사망자의 수를 확인
# index 1의 총사망자 - index 0의 총사망자 => index 1의 일일사망자

df.loc[1, '총사망자'] - df.loc[0, '총사망자']
df.loc[2, '총사망자'] - df.loc[1, '총사망자']
# ...
df.loc[819, '총사망자'] - df.loc[818, '총사망자']
```

```python
datas = [0]     # 일일확진자 수를 저장할 리스트

for i in range(1, len(df), 1):                  # 데이터프레임의 길이 len(df) => 인덱스의 수 len(df.index)
    data = df.loc[i, '총사망자'] - df.loc[i-1, '총사망자']
    datas.append(data)
```

```python
datas
```

```python
len(datas)
```

```python
df['일일사망자'] = datas
```

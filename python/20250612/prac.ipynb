#### 네이버증권 삼성전자(005920) 종합정보 크롤링 
- https://finance.naver.com/item/main.naver?code=005930

```python
codes = ['005930', '000270', '005380', '042660', '225570']
```

```python
code = codes[3]
base_url = 'https://finance.naver.com/item/main.naver?code='
res = requests.get(
    base_url + code
)
```

```python
# 응답 데이터를 문자형으로 받아 BeautifulSoup class로 파싱
soup = bs(res.text, 'html.parser')
soup
```

```python
# 파일 이름으로 사용할 종목명 추출
# div 태그 중 class가 h_company인 태그 선택
div_data = soup.find(
    'div', attrs={
        'class' : 'h_company'
    }
)
```

```python
# div_data에서 h2 태그를 선책해 문자 추출
company_name = div_data.find('h2').get_text().rstrip()
company_name
```

```python
# soup에서 div 중 class가 invest_trend인 태그 선택
invest_data = soup.find(
    'div', attrs={
        'class' : 'invest_trend'
    }
)
```

```python
dfs = pd.read_html(str(invest_data))
```

```python
dfs[0]
```

```python
# 결측치가 포함된 행 제거
trade_info = dfs[0].dropna(axis=0).reset_index(drop=True)   # axis=0 -> 행을 지우겠다 (axis='rows')
fori_org = dfs[1].dropna(axis=0).reset_index(drop=True)
```

```python
fori_org
```

```python
# csv 파일 생성
trade_info.to_csv(f'data/{company_name}_거래원정보.csv', index=False)
fori_org.to_csv(f'data/{company_name}_외국인,기관.csv', index=False)
```

- codes 리스트에 있는 종목코드에 해당하는 주식 종목의 이름, 종합정보 테이블을 불러와 csv 파일로 저장

```python
for code in codes:
    base_url = 'https://finance.naver.com/item/main.naver?code='
    res = requests.get(
        base_url + code
    )
    # 응답 데이터를 문자형으로 받아 BeautifulSoup class로 파싱
    soup = bs(res.text, 'html.parser')
    # 파일 이름으로 사용할 종목명 추출
    # div 태그 중 class가 h_company인 태그 선택
    div_data = soup.find(
        'div', attrs={
            'class' : 'h_company'
        }
    )
    # div_data에서 h2 태그를 선택해 문자 추출
    company_name = div_data.find('h2').get_text().rstrip()

    # 종합정보 데이터 추출
    # soup에서 div 중 class가 invest_trend인 태그 선택
    invest_data = soup.find(
        'div', attrs={
            'class' : 'invest_trend'
        }
    )
    dfs = pd.read_html(str(invest_data))

    # 결측치가 포함된 행 제거
    trade_info = dfs[0].dropna(axis=0).reset_index(drop=True)
    fori_org = dfs[1].dropna(axis=0).reset_index(drop=True)

    trade_info.to_csv(f'data/{company_name}_거래원정보.csv', index=False)
    fori_org.to_csv(f'data/{company_name}_외국인,기관.csv', index=False)
```

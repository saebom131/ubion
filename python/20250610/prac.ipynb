#### openAPI 수집
- 국토교통부_아파트 매매 실거래가 자료
- xml 파일
- https://www.data.go.kr/data/15126469/openapi.do


```python
import requests
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
# service_key 값이 외부에 노출되지 않음
service_key = os.getenv('service_key2')
url = 'https://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade'
params = {
    'LAWD_CD' : 41430,
    # LAWD_CD -> 지역코드, 각 지역별 코드 행정표준코드관리시스템(www.code.go.kr)의 법정동코드 10자리 중 앞 5자리  
    'DEAL_YMD' : 202403,
    # DEAL_YMD -> 계약월, 실거래 자료의 계약년월(6자리)
    'serviceKey' : service_key
}
```

```python
# 요청 보내는 부분
response = requests.get(url, params=params)
```

```python
print(response)
```

```python
print(response.content)
```
